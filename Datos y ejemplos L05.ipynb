```python
class Manzana():
    def __init__(self, df, gdf):
        self.df_aux = df
        self.df_out = gdf
        self.iniciar()
        self.habitacional()
        self.comercio()
        self.salud()
        self.educacion()
        self.deporte()

    def iniciar(self):
        df_filtrado = self.df_aux
        df_agrupado = df_filtrado.groupby('Manzana').agg(m2_Total=('Superficie (m2)','sum')).reset_index()
        df_agrupado.rename(columns={'Manzana': 'MANZ_SII'}, inplace=True)
        self.df_out = pd.merge(self.df_out, df_agrupado, on='MANZ_SII', how='left')

    def habitacional(self):    
        df_filtrado = self.df_aux[self.df_aux['Código de destino'] == 'H']
        df_agrupado = df_filtrado.groupby('Manzana').agg(m2_habitacional=('Superficie (m2)','sum')).reset_index()
        df_agrupado.rename(columns={'Manzana': 'MANZ_SII'}, inplace=True)
        self.df_out = pd.merge(self.df_out, df_agrupado, on='MANZ_SII', how='left')

    def comercio(self):
        df_filtrado = self.df_aux[self.df_aux['Código de destino'] == 'C']
        df_agrupado = df_filtrado.groupby('Manzana').agg(m2_somercio=('Superficie (m2)','sum')).reset_index()
        df_agrupado.rename(columns={'Manzana': 'MANZ_SII'}, inplace=True)
        self.df_out = pd.merge(self.df_out, df_agrupado, on='MANZ_SII', how='left')

    def salud(self):
        df_filtrado = self.df_aux[self.df_aux['Código de destino'] == 'S']
        df_agrupado = df_filtrado.groupby('Manzana').agg(m2_salud=('Superficie (m2)','sum')).reset_index()
        df_agrupado.rename(columns={'Manzana': 'MANZ_SII'}, inplace=True)
        self.df_out = pd.merge(self.df_out, df_agrupado, on='MANZ_SII', how='left')
    
    def educacion(self):
        df_filtrado = self.df_aux[self.df_aux['Código de destino'] == 'E']
        df_agrupado = df_filtrado.groupby('Manzana').agg(m2_educacion=('Superficie (m2)','sum')).reset_index()
        df_agrupado.rename(columns={'Manzana': 'MANZ_SII'}, inplace=True)
        self.df_out = pd.merge(self.df_out, df_agrupado, on='MANZ_SII', how='left')
    
    def deporte(self):
        df_filtrado = self.df_aux[self.df_aux['Código de destino'] == 'D']
        df_agrupado = df_filtrado.groupby('Manzana').agg(m2_deporte=('Superficie (m2)','sum')).reset_index()
        df_agrupado.rename(columns={'Manzana': 'MANZ_SII'}, inplace=True)
        self.df_out = pd.merge(self.df_out, df_agrupado, on='MANZ_SII', how='left')

 
a = Manzana(df_lineas_valdivia, gdf_mz_valdivia)
```

```python
import numpy as np
import matplotlib.pyplot as plt


def f_acc_caminata(d):
    # Parámetros
    k = 0.01  # Pendiente (ajusta este valor para cambiar la rapidez del decaimiento)
    t0 = 250  # Punto donde la pendiente es más pronunciada

    return 1 - (1 / (1 + np.exp(-k * (d - t0))))

power = -0.0005
f_acc_veh = lambda x: np.exp(power * x)


d = np.linspace(0, 10000, 500)
y = f_acc_caminata(d)
z = f_acc_veh(d)


plt.plot(d, y, label='Accesibilidad Caminata')
plt.plot(d, z, label='Accesibilidad Vehículo')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.xlim(0, 10000)  
plt.ylim(0, 1)      
plt.legend()
plt.grid(True)
plt.show()

```

```python
import osmnx as ox
import matplotlib.pyplot as plt
import networkx as nx
from pyproj import Proj, transform
import math

proj_utm = Proj('EPSG:32718')
proj_wgs84 = Proj('EPSG:4326')


class Valdivia():
    def __init__(self, valdivia):
        self.net = valdivia
        self.graph = ox.graph_from_place("Valdivia, Chile", network_type='drive')
        self.matriz_distancias = None
        

    def distancia(self, id_origen, id_destino):
        origen = self.net.loc[self.net['MANZ_SII'] == id_origen]['centroide'].iloc[0]
        destino = self.net.loc[self.net['MANZ_SII'] == id_destino]['centroide'].iloc[0]
        
        longitude_o, latitude_o = transform(proj_utm, proj_wgs84, origen.x, origen.y)
        longitude_d, latitude_d = transform(proj_utm, proj_wgs84, destino.x, destino.y)

        # orig_node = ox.distance.nearest_nodes(self.graph, latitude_o, longitude_o)
        # dest_node = ox.distance.nearest_nodes(self.graph, latitude_d, longitude_d)


        lat1_rad = math.radians(latitude_o)
        lon1_rad = math.radians(longitude_o)
        lat2_rad = math.radians(latitude_d)
        lon2_rad = math.radians(longitude_d)

        # Calcular la distancia euclidiana
        distancia = math.sqrt((lon2_rad - lon1_rad) ** 2 + (lat2_rad - lat1_rad) ** 2)
        
        # Escalar la distancia (en grados) a metros (aproximado)
        # 1 grado de latitud es aproximadamente 111.32 km
        # 1 grado de longitud varía según la latitud, pero se puede aproximar
        distancia_metros = distancia * 111320  # Aproximación de metros por grado de longitud a la latitud de Santiago (~-33.46)




        # d = nx.shortest_path_length(self.graph, orig_node, dest_node, weight='length')

        return distancia_metros

    def f_acc_caminata(self, d):
        # Parámetros
        k = 0.01  # Pendiente (ajusta este valor para cambiar la rapidez del decaimiento)
        t0 = 250  # Punto donde la pendiente es más pronunciada
        return 1 - (1 / (1 + np.exp(-k * (d - t0))))

    def f_acc_veh(self, x):
        return np.exp(-0.0005 * x)
    
    def generar_distancias(self):
        
        manzanas_df = self.net

        # Obtener los valores únicos de MANZ_SII
        unique_manzanas = manzanas_df['MANZ_SII'].unique()
        n = len(unique_manzanas)

        # Crear una matriz de distancias con dimensiones adecuadas
        matriz_distancias = pd.DataFrame(0, index=unique_manzanas, columns=unique_manzanas)

        # Calcular las distancias entre todas las manzanas
        for i in range(len(manzanas_df)):
            for j in range(len(manzanas_df)):
                if i != j:  # No calcular la distancia de una manzana a sí misma
                    distancia = manzanas_df['centroide'].iloc[i].distance(manzanas_df['centroide'].iloc[j])
                    matriz_distancias.loc[manzanas_df['MANZ_SII'].iloc[i], manzanas_df['MANZ_SII'].iloc[j]] = distancia

        # Mostrar la matriz de distancias
        self.matriz_distancias = matriz_distancias


        
    
    def acc_salud(self, modo, celda):
        if modo == 'cam':
            self.net['aux'] = self.net.apply(lambda row: row['m2_salud']*self.f_acc_caminata(self.distancia(celda, row['MANZ_SII'])), axis = 1)
        elif modo == 'veh':
            self.net['aux'] = self.net.apply(lambda row: row['m2_salud']*self.f_acc_veh(self.distancia(celda, row['MANZ_SII'])), axis = 1)

        acc_total = self.net['aux'].sum()
        
        return acc_total

    def acc_salud_matriz(self, modo, celda, matriz_distancias):
        # Obtener el ID de la celda actual
        celda_id = celda['MANZ_SII']
        
        # Obtener la distancia de la celda a todas las manzanas
        distancias_a_celda = matriz_distancias.loc[celda_id]
        
        # Filtrar manzanas dentro de un rango de distancia (por ejemplo, 900 metros)
        manzanas_cercanas = distancias_a_celda[distancias_a_celda < 900].index
        
        # Calcular accesibilidad para cada manzana cercana
        if modo == 'cam':
            self.net['aux'] = self.net.loc[self.net['MANZ_SII'].isin(manzanas_cercanas), 'm2_salud'] * self.f_acc_caminata(distancias_a_celda[manzanas_cercanas])
        elif modo == 'veh':
            self.net['aux'] = self.net.loc[self.net['MANZ_SII'].isin(manzanas_cercanas), 'm2_salud'] * self.f_acc_veh(distancias_a_celda[manzanas_cercanas])

        # Sumar la accesibilidad total
        acc_total = self.net['aux'].sum()

        return acc_total


ciudad = Valdivia(manzanas_valdivia)
```

```python
import geopandas as gpd

# Cargar los shapes del SII
sii_gdf = manzanas_valdivia

# Cargar el grid
grid_gdf = celdas_valdivia
# Transformar el CRS del grid al CRS de SII (EPSG:4326)
sii_gdf = sii_gdf.to_crs(epsg=4326)
grid_gdf = grid_gdf.to_crs(epsg=4326)

# Calcular la intersección
interseccion_gdf = gpd.overlay(sii_gdf, grid_gdf, how='intersection')

# Calcular el área de intersección
interseccion_gdf['area_interseccion'] = interseccion_gdf.geometry.area

# Calcular el área total de cada manzana del SII
sii_gdf['area_total_manzana'] = sii_gdf.geometry.area

# Crear una columna en intersección para almacenar la proporción
interseccion_gdf['proporcion_area'] = interseccion_gdf['area_interseccion'] / interseccion_gdf['area_total_manzana']

# Agregar la proporción al GeoDataFrame original del SII
sii_gdf = sii_gdf.merge(
    interseccion_gdf[['cell', 'MANZ_SII', 'proporcion_area']],
    on='MANZ_SII',
    how='left'
)

sii_gdf = sii_gdf.drop(columns=['area_total_manzana'])
sii_gdf = sii_gdf.drop_duplicates(subset=['MANZ_SII', 'cell'])

sii_gdf

```

```python
sii_gdf['m2_hab'] = sii_gdf['proporcion_area']*sii_gdf['m2_habitacional']
sii_gdf['m2_com'] = sii_gdf['proporcion_area']*sii_gdf['m2_somercio']
sii_gdf['m2_sal'] = sii_gdf['proporcion_area']*sii_gdf['m2_salud']
sii_gdf['m2_edu'] = sii_gdf['proporcion_area']*sii_gdf['m2_educacion']
sii_gdf['m2_dep'] = sii_gdf['proporcion_area']*sii_gdf['m2_deporte']
gdf_agrupado = sii_gdf.groupby('cell', as_index=False).agg({'m2_hab': 'sum', 'm2_com': 'sum', 'm2_sal': 'sum', 'm2_edu': 'sum', 'm2_dep': 'sum'})
gdf_agrupado
```
